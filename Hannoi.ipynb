```python
from tHanoiAbstract import tHanoiAbstract
import sys


class Hannoi(tHanoiAbstract):
    """
    :Description: This class implements the Tower of Hanoi algorithm using recursion.

    :parameters:
        disks(int)      - Number of disks
        source(str)     - Source rod
        auxiliary(str)  - Auxiliary rod
        target(str)     - Target rod

    **Terminal command**

    Format:

        python3 Hannoi.py <numberOfDisks>

    Example Usage:

        python3 Hannoi.py 3
    """

    def move_disk(self, source, target):
        """
        Prints the movement of a disk.

        :param source:
        :param target:
        :return:
        """
        print(f"Move disk from {source} to {target}")

    def solve(self, disks, source, auxiliary, target):
        """
        Solves Tower of Hanoi recursively.

        :param disks:
        :param source:
        :param auxiliary:
        :param target:
        :return:
        """
        if disks == 1:
            self.move_disk(source, target)
            return

        self.solve(disks - 1, source, target, auxiliary)
        self.move_disk(source, target)
        self.solve(disks - 1, auxiliary, source, target)


if __name__ == "__main__":

    if len(sys.argv) != 2:
        print("Error: Syntax is python3 Hannoi.py numberOfDisks")
    else:
        disks = int(sys.argv[1])

        hanoi = Hannoi()
        hanoi.solve(disks, "A", "B", "C")
```
