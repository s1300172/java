```python
from pigeonHoleSortAbstract import pigeonHoleSortAbstract
import random
import sys


class pigeonHoleSort(pigeonHoleSortAbstract):
    """
    :Description: This class implements the Pigeon Hole Sorting algorithm.

    :parameters:
        arr(list) - List of integers to be sorted

    **Terminal command**

    Format:

        python3 pigeonHoleSort.py <randomN>

    Example Usage:

        python3 pigeonHoleSort.py 100
    """

    def sort(self, arr):
        """
        Implements pigeon hole sorting.

        :param arr:
        :return:
        """

        minimum = min(arr)
        maximum = max(arr)

        size = maximum - minimum + 1

        holes = [0] * size

        for value in arr:
            holes[value - minimum] += 1

        sorted_arr = []

        for i in range(size):
            while holes[i] > 0:
                sorted_arr.append(i + minimum)
                holes[i] -= 1

        return sorted_arr


if __name__ == "__main__":

    if len(sys.argv) != 2:
        print(
            "Error: The user does not specify the correct number of input parameters."
        )
        print("Syntax is: python3 pigeonHoleSort.py randomN")

    else:

        n = int(sys.argv[1])

        arr = [random.randint(0, 1000) for _ in range(n)]

        print("Original List:")
        print(arr)

        sorter = pigeonHoleSort()

        print("\nSorted List:")
        print(sorter.sort(arr))
```
